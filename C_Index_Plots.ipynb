```python
image_paths = glob.glob('results/kidney_emd_cens_kidney_2022_01_11_12_13_27/*/Figures/c_index_*')

fig, axs = plt.subplots(4,5, figsize=(30,20))
for i, im_path in enumerate(image_paths):
    im = cv2.cvtColor(cv2.imread(im_path), cv2.COLOR_BGR2RGB)
    axs[i//5,i%5].imshow(im)
    axs[i//5,i%5].set_title(im_path.split('/')[2])
    axs[i//5,i%5].set_xticks([])
    axs[i//5,i%5].set_yticks([])
```
